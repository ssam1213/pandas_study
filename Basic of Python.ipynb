# 6. Pandas 고급 두 DataFrame 병합하기


```python
df1 = pd.DataFrame({
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
},index=[0,1,2,3])
df2 = pd.DataFrame({
    'A':['A4','A5','A6','A7'],
    'B':['B4','B5','B6','B7'],
    'C':['C4','C5','C6','C7'],
    'D':['D4','D5','D6','D7']
},index=[4,5,6,7])
df3 = pd.DataFrame({
    'A':['A8','A9','A10','A11'],
    'B':['B8','B9','B10','B11'],
    'C':['C8','C9','C10','C11'],
    'D':['D8','D9','D10','D11']
},index=[8,9,10,11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
result = pd.concat([df1,df2,df3])
```

```python
result
```

concat함수 열방향으로 단순히 합침


```python
result = pd.concat([df1,df2,df3],keys=['x','y','z'])
```

```python
result
```

```python
result.index
```

```python
result.index.get_level_values(0)
```

```python
result.index.get_level_values(1)
```

```python
df4 = pd.DataFrame({
    'B':['B2','B3','B6','B7'],
    'D':['D2','D3','D6','D7'],
    'F':['F2','F3','F6','F7']
},index=[2,3,6,7])
result = pd.concat([df1,df4],axis=1)
```

```python
df1
```

```python
df4
```

```python
result
```

concat은 index기준으로 데이터를 합치기 때무에 데이터가 올 수 없는 인덱스에는 NaN

```python
result = pd.concat([df1,df4],axis=1,join='inner')
result
```

공통된 인덱스는 합치고 아닌 인덱스는 버림

```python
result = pd.concat([df1,df4],axis=1,join_axes=[df1.index])
result
```

df1의 인덱스에 맞추도록


```python
result = pd.concat([df1,df4],ignore_index=False)
result
```

책은 ignore_index=True인데 sort=True사용해서 

```python
result = pd.concat([df1,df4],sort=False)
result
```

```python
left = pd.DataFrame({
    'key':['K0','K4','K2','K3'],
    'A':['A0','A1',"A2",'A3'],
    'B':['B0','B1','B2','B3']
})
right = pd.DataFrame({
    'key':['K0','K1','K2','K3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on='key')
```

```python
pd.merge(left,right,how='left',on='key')
```

```python
pd.merge(left,right,how='right',on='key')
```

```python
pd.merge(left,right,how='outer',on='key')
```
